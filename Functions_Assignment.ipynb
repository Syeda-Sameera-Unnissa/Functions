1. What is the difference between a function and a method in Python?
---
The key differences between a **function** and a **method** in Python are:

1. **Association**:
   - **Function**: Independent and not tied to any object. Defined globally or within modules.
   - **Method**: Associated with an object or class. Defined inside a class.

2. **Calling Context**:
   - **Function**: Called directly using its name.
    
     def greet(name):
         return f"Hello, {name}!"
     greet("Alice")
     
   - **Method**: Called on an instance or class (e.g., `object.method()`).
     
     class Greeter:
         def greet(self, name):
             return f"Hello, {name}!"
     greeter = Greeter()
     greeter.greet("Alice")

3. **First Argument**:
   - **Function**: Arguments are explicitly passed by the caller.
   - **Method**: Automatically receives the instance (`self`) or class (`cls`) as the first argument.

In essence, a **method is a function defined within a class and operates on objects of that class**, while a **function is a standalone block of code**.

4. What is the purpose of the 'return' statement in a Python function?
---
The **`return` statement** in a Python function is used to:

1. **Send a value back to the caller**: It specifies the result of a function after its execution.
   
   def add(a, b):
       return a + b

   result = add(3, 5)  # 'result' will hold the value 8
   

2. **Terminate the function**: Once the `return` statement is executed, the function exits immediately.
   
   def check_positive(num):
       if num > 0:
           return "Positive"
       return "Non-positive"  # Stops here if num <= 0
  

3. **Return multiple values** (packed as a tuple):
   
   def stats(a, b):
       return a + b, a * b  # Returns a tuple

   sum_, product = stats(3, 4)  # sum_ = 7, product = 12
  

If no `return` statement is provided, the function returns `None` by default.

5. What are iterators in Python and how do they differ from iterables?
---
**Iterators**:
- Objects that allow sequential access to elements, one at a time.
- Must implement `__iter__()` and `__next__()`.
- Cannot be reused once exhausted.

**Iterables**:
- Objects that can return an iterator using `__iter__()`.

- Can create a new iterator for each iteration.

**Key Difference**:
- **Iterable**: The collection (e.g., list, string) that can be iterated.
- **Iterator**: The object created from an iterable to fetch items one by one.

6. Explain the concept of generators in Python and how ar they defined?
---
**Generators in Python**:
- **Definition**: Generators are special functions that produce a sequence of values lazily, one at a time, using the `yield` keyword.
- **Purpose**: They allow efficient iteration over large datasets without storing the entire data in memory.

---

**How Are Generators Defined?**
1. **Using a Function with `yield`**:
   - Instead of `return`, use `yield` to produce a value and pause the functionâ€™s state.
   
   def my_generator():
       for i in range(3):
           yield i

2. **Using Generator Expressions**:
   - Similar to list comprehensions but with parentheses.
   
   gen = (x**2 for x in range(5))

---

**Key Characteristics**:
- **State Preservation**: Generators maintain their state between calls.
- **Lazy Evaluation**: Values are generated only when needed.
- **Memory Efficiency**: Ideal for large data processing.

**Example**:

def count_up_to(n):
    for i in range(1, n + 1):
        yield i

gen = count_up_to(3)
print(next(gen))  # Output: 1
print(next(gen))  # Output: 2


7. What are the advantages of using generators over regular functions?
---
**Advantages of Generators over Regular Functions**:

1. **Memory Efficiency**:
   - Generators yield items one at a time, avoiding the need to store entire datasets in memory.
   - Ideal for large datasets or streams of data.

2. **Lazy Evaluation**:
   - Values are generated only when needed, improving performance and reducing overhead.

3. **Simpler Code**:
   - Generators simplify writing iterable code compared to manually implementing iterators.

4. **State Preservation**:
   - Generators automatically save their state between `yield` calls, making them easy to pause and resume.

5. **Improved Performance**:
   - No need to compute all values upfront, which speeds up execution for cases where not all values are needed.

Example:

def generate_numbers(n):
    for i in range(n):
        yield i

gen = generate_numbers(5)
print(next(gen))  # Generates only when called

8. What is a lambda function in Python and when is it typically used?
---
**Lambda Function in Python**:
- A **lambda function** is an anonymous, single-line function defined using the `lambda` keyword.
- Syntax: `lambda arguments: expression`

**When to Use Lambda Functions**:
- When a small, throwaway function is needed for a short period.
- Typically used in:
  - **Functional Programming**: With functions like `map()`, `filter()`, `reduce()`.
  - **Sorting**: As a custom key in `sorted()`.
  - **Callbacks**: To pass a quick inline function.

**Example**:
Using lambda with map()
numbers = [1, 2, 3]
squares = map(lambda x: x**2, numbers)
print(list(squares))  # Output: [1, 4, 9]

Lambda functions are concise and avoid the need for explicitly defining named functions.

10. What is the difference between 'map()', 'reduce()' and 'filter()' functions in Python?
---
The **`map()`**, **`reduce()`**, and **`filter()`** functions in Python serve distinct purposes in functional programming and operate on iterables.

The **`map()`** function applies a given function to each element of an iterable, transforming the elements and returning a new iterable. For example, squaring each number in a list:

nums = [1, 2, 3, 4]
result = map(lambda x: x**2, nums)  # Squares each number
print(list(result))  # Output: [1, 4, 9, 16]

The **`reduce()`** function (from the `functools` module) reduces an iterable to a single cumulative value by repeatedly applying a function to its elements. For example, calculating the sum of all numbers in a list:

from functools import reduce
nums = [1, 2, 3, 4]
result = reduce(lambda x, y: x + y, nums)  # Sums all numbers
print(result)  # Output: 10

The **`filter()`** function extracts elements from an iterable that satisfy a given condition, returning only those that meet the criteria.

For example, filtering out even numbers from a list:

nums = [1, 2, 3, 4]
result = filter(lambda x: x % 2 == 0, nums)  # Filters even numbers
print(list(result))  # Output: [2, 4]

In summary, **`map()`** transforms elements, **`reduce()`** aggregates them, and **`filter()`** isolates specific elements based on a condition.

```python
#1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

def sum_of_even_numbers(numbers):
    even_sum = 0
    for num in numbers:
        if num % 2 == 0:
            even_sum += num
    return even_sum

input_numbers = input("Enter a list of numbers separated by comma: ")

numbers = list(map(int, input_numbers.split(',')))

result = sum_of_even_numbers(numbers)
print("The sum of all even numbers is:", result)
```

```python
#2. Create a Python function that accepts a string and returns the reverse of that string.

def reverse_string(input_string):
    """
    Reverse the given string.

    Parameters:
        input_string (str): The string to reverse.

    Returns:
        str: The reversed string.
    """
    reversed_str = ""
    for char in input_string:
        reversed_str = char + reversed_str
    return reversed_str

# Example usage
if __name__ == "__main__":
    user_input = input("Enter a string: ")
    result = reverse_string(user_input)
    print("The reversed string is:", result)
```

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime(n):

    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Generate a list of all primes from 1 to 200
primes = []
for num in range(1, 201):
    if is_prime(num):
        primes.append(num)

# Print all primes from 1 to 200
print("Prime numbers from 1 to 200:")
print(primes)
```

```python
#5. Create an iterator class in Python that generates the fibonacci sequence up to a specified number of terms.

class FibonacciIterator:

    def __init__(self, n_terms):

        self.n_terms = n_terms
        self.current_term = 0
        self.a, self.b = 0, 1  # Initialize the first two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_term >= self.n_terms:
            raise StopIteration
        if self.current_term == 0:
            self.current_term += 1
            return self.a
        elif self.current_term == 1:
            self.current_term += 1
            return self.b
        else:
            self.current_term += 1
            self.a, self.b = self.b, self.a + self.b
            return self.b

# Example usage:
n = 10  # Number of terms in the Fibonacci sequence
fib_iter = FibonacciIterator(n)

print(f"Fibonacci sequence up to {n} terms:")
for num in fib_iter:
    print(num, end=" ")
```

```python
#6. Write a generator function in Python that yields the power of 2 up to a given exponent.

def powers_of_two(max_exponent):
    for exponent in range(max_exponent + 1):
        yield 2 ** exponent

max_exp = 10  # Maximum exponent
print(f"Powers of 2 up to 2^{max_exp}:")
for power in powers_of_two(max_exp):
    print(power, end=" ")
```

```python
#7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Removes any leading/trailing whitespace, including newlines

# Example usage:
file_path = 'example.txt'  # Replace with the path to your file
for line in read_file_line_by_line(file_path):
    print(line)
```

```python
#9. Write a Python program that uses 'map()' to convert a list of temperatures from Celsius to Fahrenheit.

# Function to convert Celsius to Fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

# List of temperatures in Celsius
celsius_temperatures = [0, 20, 37, 100]

# Convert to Fahrenheit using map()
fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))

# Print the results
print("Temperatures in Celsius:", celsius_temperatures)
print("Temperatures in Fahrenheit:", fahrenheit_temperatures)
```

```python
#10. Create a Python program that uses 'filter()' to remove all the vowels from a given string.

# Function to check if a character is not a vowel
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

# Input string
input_string = "This is a sample string with vowels."

# Remove vowels using filter()
filtered_string = ''.join(filter(is_not_vowel, input_string))

# Print the result
print("Original string:", input_string)
print("String without vowels:", filtered_string)
```

```python
#11(a). Write a Python program, which returns a list with 2 tuples. Each tuple consists of the order number and the product of the price per item and the quantity. The product should be increased by 10,-â‚¬ if the value of the order is smaller than 10,000â‚¬.

# Data: List of sublists containing order details
orders = [
    [34587, "Learning Python, Mark Lutz", 4, 40.95],
    [98762, "Programming Python, Mark Lutz", 5, 56.80],
    [77226, "Head First Python, Paul Barry", 3, 32.95],
    [88112, "Einfuhrung in Python 3, Bernd Klein", 3, 24.99]
]

# Function to calculate the total cost for each order with the condition
def process_orders(order_list):
    result = []
    for order in order_list:
        order_num = order[0]
        quantity = order[2]
        price_per_item = order[3]
        total_cost = quantity * price_per_item
        # Add 10â‚¬ if the total cost is less than 100.00â‚¬
        if total_cost < 100:
            total_cost += 10
        result.append((order_num, round(total_cost, 2)))  # Rounded to 2 decimal places
    return result

# Process the orders and print the result
processed_orders = process_orders(orders)
print(processed_orders)
```
